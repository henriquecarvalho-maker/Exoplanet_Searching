### O que são Exoplanetas ?
<p>Exoplanetas são copos celestes planetários que não fazem parte do nosso sistema solar, mas sim de outros sistemas estelares. O estudo desses corpos é um passo sem precedentes para a procura de vida em nosso universo.</p>
<p>Exitem diversas variáveis para que a vida floresça: Esse planeta precisa ser <b>Rochoso</b>, esta na <b>Green Zone</b>, o núcleo precisa ser ativo para que o planeta tenha um <b>Campo Magnético</b> e atividade geologica, a atmosfera precisa ser densa na medida certa para que tenha <b>Efeito Estufa Natural</b>, é necessário um <b>Satélite Natural</b> para que haja uma estabilidade climática, no mesmo sistema solar seria ideal haver um outro <b>Corpo Planetário Super Massivo</b> semelhante a Saturno ou Júpiter para prover uma proteção contra eventuais impactos.</p>

## Dados de Exoplanetas Descorbetos Até Então
<hr>
<p>A nossa busca é pelo os Terrestres e como podemos ver abaixo eles são raros, em 5250 encontrados apenas 195 são terrestres.</p>

```python
ax = sns.countplot(x='planet_type', data=dados)
ax.set_title('Tipos dos Exeplanetas Econtrados')
ax.set_xlabel('Tipos Planetarios')
ax.set_ylabel('Numero de Planetas')
```

```python
numeros_de_plents = dados['name'].count()
numeros_de_plents_terrestres = dados.query('planet_type == "Terrestrial"')
numeros_de_plents_terrestres_ = numeros_de_plents_terrestres['name'].count()

print(f'Planetas descodebetos: {numeros_de_plents}')
print(f'Planetas Terrestre:{numeros_de_plents_terrestres_}')
```

```python
expl_terrestris = dados.query('planet_type == "Terrestrial"')
expl_terrestris.head()
```

```python
ax = sns.scatterplot(x='distance', y='planet_type', data=dados)
ax.set_title('Distancia dos Planetas em Relação a Terra')
ax.set_xlabel('Distancia em Anos Luz')
ax.set_ylabel('Planetas')
```
