### O que são Exoplanetas ?
<p>Exoplanetas são copos celestes planetários que não fazem parte do nosso sistema solar, mas sim de outros sistemas estelares. O estudo desses corpos é um passo sem precedentes para a procura de vida em nosso universo.</p>
<p>Exitem diversas variáveis para que a vida floresça: Esse planeta precisa ser <b>Rochoso</b>, esta na <b>Green Zone</b>, o núcleo precisa ser ativo para que o planeta tenha um <b>Campo Magnético</b> e atividade geológica, a atmosfera precisa ser densa na medida certa para que tenha <b>Efeito Estufa Natural</b>, é necessário um <b>Satélite Natural</b> para que haja uma estabilidade climática, no mesmo sistema solar seria ideal haver um outro <b>Corpo Planetário Super Massivo</b> semelhante a Saturno ou Júpiter para prover uma proteção contra eventuais impactos.</p>

<hr>
<p>Sabemos que, umas das mais famosas leis da física diz que qualquer objeto que contem massa <b>Não</b> pode viajar a velocidade da luz, pois teria que adquirir massa infinita para alcança-la. Exitem algumas tentativas em solucionar esse problema e tentar contorna-lo, tentativas teóricas como a <b>Propulsão Alcubierre</b>, que matematicamente faz sentido, no entanto quando posta fisicamente acaba sendo impraticável.</p>

<p>Na teoria essa solução é elegante, os cálculos propõem <em>"Um método de alongamento do espaço em uma onda que, em teoria, poderia fazer com que o tecido do espaço à frente de uma nave espacial se contraia, enquanto que o tecido que está atrás da nave se expanda. A nave se deslocaria "surfando" esta onda dentro de uma região conhecida como <b>bolha de dobra</b>."</em> Dessa forma contornando o problema das distâncias astronômicas, mas também uma coisa legal acontece, os passageiros não são afetados pelos efeitos relativísticos previsto por Einstein como a dilatação temporal, em teoria a nave não se move e sim o espaço ao redor dela.</p>

## Planetas na Zona Abitavel com até 150 anos luz
<hr>
<p>Em astronomia e astrobiologia, a zona habitável, ou mais precisamente a zona habitável circunstelar, é a gama de órbitas em torno de uma estrela dentro da qual uma superfície planetária pode suportar água líquida, dada pressão atmosférica suficiente.</p>
