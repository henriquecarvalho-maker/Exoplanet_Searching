### O que são Exoplanetas ?
<p>Exoplanetas são copos celestes planetários que não fazem parte do nosso sistema solar, mas sim de outros sistemas estelares. O estudo desses corpos é um passo sem precedentes para a procura de vida em nosso universo.</p>
<p>Exitem diversas variáveis para que a vida floresça: Esse planeta precisa ser <b>Rochoso</b>, esta na <b>Green Zone</b>, o núcleo precisa ser ativo para que o planeta tenha um <b>Campo Magnético</b> e atividade geologica, a atmosfera precisa ser densa na medida certa para que tenha <b>Efeito Estufa Natural</b>, é necessário um <b>Satélite Natural</b> para que haja uma estabilidade climática, no mesmo sistema solar seria ideal haver um outro <b>Corpo Planetário Super Massivo</b> semelhante a Saturno ou Júpiter para prover uma proteção contra eventuais impactos.</p>

# Distância 
<hr>
<p>A distância é um grande problema para nós, viajar grandes distâncias é um desafio que talvez no futuro possamos superar. O objeto mais rápido criado pelo o ser humano foi a sonda New Horizon que viajava no vaco a 21km por segundo ou mais de 75 mil km/h, ela levou 10 meses para chegar em Júpiter.</p>

```python
expl_terrestris = dados.query('planet_type == "Terrestrial"')
expl_terrestris.head(3)
```

```python
ax = sns.scatterplot(x='distance', y='planet_type', data=dados)
ax.set_title('Distancia dos Planetas em Relação a Terra')
ax.set_xlabel('Distancia em Anos Luz')
ax.set_ylabel('Planetas')
```

<p>Sabemos que, de acordo a umas das mais famosas leis da física que diz que qualquer objeto que contem massa <b>Não</b> pode viajar a velocidade da luz, pois teria que ganhar massa infinita para alcança-la. Exitem algumas tentativas em solucionar esse problema e tentar contorna-lo, tentativas teoricas como a <b>Propulsão Alcubierre</b>, que matematicamente faz sentido no entanto quando posta fisicamente acaba sendo impraticavel.</p>
<p>Na teoria essa solução é elegante, os calculos propõem <em>"Um método de alongamento do espaço em uma onda que, em teoria, poderia fazer com que o tecido do espaço à frente de uma nave espacial se contraia, enquanto que o tecido que está atrás da nave se expanda. A nave se deslocaria "surfando" esta onda dentro de uma região conhecida como <b>bolha de dobra</b>."</em> Dessa forma contornando o problema das distancias astronomicas, mas tambem uma coisa legal acontece, os passageiros não são afetados pelos efeitos relativisticos previsto por Einsten como a dilatação temporal, em teoria a nave não se move e sim o espaço ao redor dela.</p>
<img width="520" height="405" src='Imagens/Alcubierre.png'>

```python
fig = plt.figure(figsize=(10,5))
ax = sns.scatterplot(x='distance', y='discovery_year', hue='planet_type', style='planet_type', data=dados)
ax.set_ylabel('Ano de Descoberta')
ax.set_xlabel('Distancia')
```

### Planetas com até 20 anos Luz de Distância

```python
planet_10ly = dados.query('distance <= 20' )
planet_10ly.head()
```

```python
planet_10ly['name'].count()
```

```python
ax = sns.countplot(x='planet_type', data=planet_10ly)
ax.set_title('Planetas com Distancias até 20 Anos Luz')
ax.set_xlabel('Tipos Planetarios')
ax.set_ylabel('Numero de Planetas')
```

<p style='font-size: 16px'> <b>Problemas e Tentativas de Soluções</b></p>
<p>Infelizmente, a <b>Propulsão Alcubierre</b> quando posto fisicamente tem problemas imensos. Para conseguir a dobra nessecitaria de quantidades de energia colossal, assim como propriedades que ainda temos pouco conhecimento como matéria exotica de hipotética massa negatica.</p> <p>Tentativas de soluções como o estudo do <b>Efeito Casimir</b> envolvendo interação com partículas virtuais são plausíveis mas muito embrionárias, consisederando tambem que essa é apenas um dos varios problemas.</p>
